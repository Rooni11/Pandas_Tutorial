<h1 style = "color:green" align="center">Handle Missing Data (Replace function)</h1>

```python
import pandas as pd
import numpy as np
df = pd.read_csv('6.weather_data.csv')
df
```

<h4 style = "color:purple">1.replacing values</h4>

```python
df1 = df.replace(-99999,np.NaN)
df1
```

```python
df2 = pd.read_csv('6.weather_data2.csv')
df2
```

```python
df3 = df2.replace([-99999,-88888,'0'],np.NaN)
df3
```

```python
df4 = df2.replace({
    'temperature': -99999,
    'windspeed': [-99999,-88888],
    'event': '0'
},np.NaN)

df4
```

```python
df5 = df2.replace({
    -99999: np.NaN,
    -88888: np.NaN,
    '0': 'Sunny'
})

df5
```

```python
d = pd.read_csv('6.weather_data3.csv')
d
```

<h4 style = "color:purple">2.Removing unwanted letter</h4> 

```python
d1 = d.replace('[a-zA-Z]','',regex=True)
d1
```

```python
d2 = d.replace({
    'temperature': '[a-zA-Z]',
    'windspeed': '[a-zA-Z]'
},'',regex=True)

d2
```

<h4 style = "color:purple">3.Removing unwanted digit</h4> 

```python
data = {'fruits': ['Apple1', 'Mango2', 'Banana3', 'Grape4']}
f = pd.DataFrame(data)
f
```

```python
# Remove digits from the 'fruits' column
f['fruits'] = f['fruits'].str.replace('\d+', '',regex=True)
#df['fruits'] = df['fruits'].str.replace('\d+', '',regex=True)
f
```

<h4 style = "color:purple">4.Encoding with replace method</h4> 

```python
r = pd.DataFrame({
    'score': ['exceptional','average','good', 'poor','average','exceptional'],
    'student': ['rob','naya','roni','tom','radin','rukaiya']
})

r
```

```python
r1 = r.replace(['poor','average','good','exceptional'],[1,2,3,4])
#r1 = r.replace({'poor':1,'average':2,'good':3,'exceptional':4})
r1
```
