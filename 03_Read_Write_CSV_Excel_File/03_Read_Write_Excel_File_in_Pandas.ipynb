### 'Converters' function to operate Column Wise:

```python
import pandas as pd

def convert_people_cell(cell):
    if cell in ['n.a.',-1]:
        return "sam walton"
    return cell

def convert_eps_cell(cell):
    if cell == 'not available':
        return None
    return cell

def convert_price_cell(cell):
    if cell == 'n.a.':
        return None
    return cell

def convert_revenue_cell(cell):
    if cell == -1:
        return None
    return cell
    
df = pd.read_excel('4.stock_data.xlsx',converters = {
    'people':convert_people_cell,
    'eps': convert_eps_cell,
    'price': convert_price_cell,
    'revenue': convert_revenue_cell
})
df
```

## 2.Write Excel File

```python
df.to_excel('new.xlsx',sheet_name='stock_data',index=False,startrow=1,startcol=3)

#startrow / startcol argument works only excel file, not csv file
```

#### Write an Excel File with 2 different sheets from 2 different DataFrame

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL','WMT','MSFT'],
    'price': [845,65,64],
    'pe': [30,25,12],
    'eps': [12.5,22.4,2.01]
})

df_stocks
```

```python
df_weather = pd.DataFrame({
    'day' : ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature' : [32,35,28],
    'windspeed': [6,7,2],
    'event' : ['Rain', 'Sunny', 'Snow']
})

df_weather
```

```python
with pd.ExcelWriter('4.new_stock_weather.xlsx') as Writer:
    df_stocks.to_excel(Writer,sheet_name='stocks',index=False)
    df_weather.to_excel(Writer,sheet_name='weather',index=False)
```
