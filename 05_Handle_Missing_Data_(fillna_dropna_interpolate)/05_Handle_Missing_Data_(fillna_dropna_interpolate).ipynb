<h1 style="color: green"> 1.Fillna()</h1>

**We will perform following:**
1. 1_1 Filling missing Value with values
2. 1_2 Filling missing Value w.r.t Columns
3. 1_3 Use 'ffill' and 'bfill' method
4. 1_4 Use fill method w.r.t Columns

<h3 style="color: purple">1.1 Filling missing Value with values</h3>

```python
df1 = df.fillna(0)
df1
```

```python
q = df.fillna(1000000)
q
```

```python
a = df.fillna(df.mean())
a
```

<h3 style="color:purple">1.2 Filling missing Value w.r.t Columns</h3>

```python
df['temperature'].mean()
```

```python
df['windspeed'].median()
```

```python
df2 = df.fillna({
    'temperature': df['temperature'].mean(),
    'windspeed': df['windspeed'].median(),
    'event': 'no event'
})

df2
```

<h3 style="color:purple">1.3 Use 'ffill' and 'bfill' method</h3>

```python
df3 = df.fillna(method='ffill')
df3
```

```python
df4 = df.fillna(method='bfill')
df4
```

**Notes:**
1. 'ffill' means forward fill, it fill the missing value by Upper Row value.
2. 'bfill' means backward fill, it fill the missing value by Lower Row value.

<h3 style="color:purple">1.4 ffill/bfill w.r.t Columns</h3> 

```python
df3 = df.fillna(method='bfill',axis='columns')
df3
```

<h3 style="color:purple">1.5 Use fill method with limit function</h3> 

```python
df4 = df.fillna(method='ffill',limit=1)
df4
```

<h1 style="color: green"> 2.Interpolate()</h1>

```python
df_pol = df.interpolate()
df_pol
```

**Notes:**
1. df.interpolate uses linear formula to fill missing value.

```python
d = pd.read_csv('5.weather_data2.csv',parse_dates=['day'])

#set day as index
d = d.set_index('day')
d
```

```python
d3 = d.interpolate(method='time')
d3
```

**Notes:**

<h1 style="color: green"> 3.dropna()</h1>

```python
#drop all rows that contain at least a null value
dd = d.dropna()
dd
```

```python
#drop specific rows that contain totally null value
dd2 = d.dropna(how='all') # default = 'any'
dd2
```

```python
dd3 = d.dropna(thresh=1)
dd3
```

```python
dd4 = d.dropna(thresh=2)
dd4
```

**Notes:**
1. dropna(thresh=n) indicate that df contain that row which has minimun n correct/valid value.
2. dropna(thresh=1) indicate that df contain that row which has minimun 1 correct/valid value.
3. practically dropna(how='all') and dropna(thresh=1) are same, it eliminate the row which totally null.
4. we can use all method w.r.t. column using axis=1/columns.
